```python
#importing the libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None#adjust the confirugation of the plot that we create


#read in the file
df = pd.read_csv('D:\Data Analysis\movies.csv')
```

```python
#lets look at the data

df.head(10)
```

```python
for col in df.columns:
    pct_missing= round(np.mean(df[col].isnull()),2)
    print('{} - {}%'.format(col, pct_missing))
```

```python
#data types for our column

df.dtypes
```

```python
df.sort_values(by=['gross'], inplace=False, ascending= False)
```

```python
# drop any duplicates

df.drop_duplicates()
```

```python
#scatter plot using budget

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Title')

plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df,  scatter_kws={"color": "blue"} , line_kws={"color": "yellow"})
```

```python
# lets look at the correlation 
```

```python
df.corr()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#lets compare the tables

df
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('corelation Matrix for numeric Features')

plt.xlabel('Movie feature')

plt.ylabel('Movie feature')

plt.show()
```
